```python
import pandas as pd
import numpy as np

df1 = pd.read_csv('testdata/1/saad_20cm_LEFT_aug.csv')
df2 = pd.read_csv('testdata/1/saad_20cm_RIGHT_aug.csv')
df3 = pd.read_csv('testdata/1/saad_20cm_UP_aug.csv')
df4 = pd.read_csv('testdata/1/saad_20cm_DOWN_aug.csv')
df5 = pd.read_csv('testdata/1/saad_20cm_CW_aug.csv')
df6 = pd.read_csv('testdata/1/saad_20cm_CCW_aug.csv')
df7 = pd.read_csv('testdata/1/saad_20cm_X_aug.csv')
df8 = pd.read_csv('testdata/1/saad_20cm_PUSH_aug.csv')
df9 = pd.read_csv('testdata/1/saad_20cm_HOLD_aug.csv')
df10 = pd.read_csv('testdata/1/saad_20cm_DPUSH_aug.csv')



df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10], axis=0)
df = df.sample(frac=1, random_state=42)
df = df.reset_index(drop=True)
df
```

```python
df["GESTURE"].value_counts()
```

```python
df.to_csv('testdata/1/dataset_ALL_saad_20cm.csv', index=False)
```
