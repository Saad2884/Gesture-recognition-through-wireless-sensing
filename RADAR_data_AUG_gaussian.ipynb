```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('testdata/1/dataset_ALL_saad_20cm.csv')
float_data = df.select_dtypes(include=np.number).values  # Select only float columns
class_labels = df['GESTURE'].values  # Extract the class column as string labels
```

```python
def augment_float_data(data, factor=2):
    augmented_data = []
    for _ in range(factor):
        augmented_data.append(data + np.random.normal(0, 0.2, size=data.shape))
    return np.concatenate(augmented_data)
```

```python
augmented_float_data = augment_float_data(float_data, factor=10)  # Generate 5 times more augmented samples
```

```python
# Repeat the class labels to match the number of augmented samples
repeated_class_labels = np.repeat(class_labels, augmented_float_data.shape[0] // class_labels.shape[0])

# Trim the repeated class labels to match the number of augmented samples exactly
repeated_class_labels = repeated_class_labels[:augmented_float_data.shape[0]]

augmented_data = np.column_stack((augmented_float_data, repeated_class_labels))
augmented_df = pd.DataFrame(augmented_data, columns=df.columns)
```

```python
augmented_df.to_csv('testdata/1/dataset_ALL_saad_20cm_big.csv', index=False)
```
