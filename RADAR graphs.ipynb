```python
lstm_sig = [87.04, 87.76, 86.59, 81.22, 70.72]
lstm_sof = [86.25, 89.45, 88.06, 84.69, 67.93]
bi_lstm_sig = [93.62, 99.23, 98.66, 98.00, 95.93]
bi_lstm_sof = [93.26, 99.45, 99.30, 98.89, 93.80]
gru_sig = [89.15, 89.32, 90.72, 90.40, 87.76]
gru_sof = [89.60, 90.00, 93.07, 92.15, 89.15]
bi_gru_sig = [95.33, 99.85, 99.53, 97.40, 93.97]
bi_gru_sof = [95.63, 99.68, 98.53, 98.08, 96.44]

layer2_acc = [lstm_sig[1], lstm_sof[1], bi_lstm_sig[1], bi_lstm_sof[1], gru_sig[1], gru_sof[1], bi_gru_sig[1], bi_gru_sof[1]]
alg = [0,1,2,3,4,5,6,7]
xticklabels = ['MLA = LSTM, AF = sigmoid', 'MLA = LSTM, AF = softmax', 'MLA = Bi-directional LSTM, AF = sigmoid', 'MLA = Bi-directional LSTM, AF = softmax','MLA = GRU, AF = sigmoid', 'MLA = GRU, AF = softmax', 'MLA = Bi-directional GRU, AF = sigmoid', 'MLA = Bi-directional GRU, AF = softmax']
plt.barh(alg,layer2_acc)
plt.yticks(alg,xticklabels)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a Seaborn style (optional)
sns.set(style="dark")

# Prepare data
categories = xticklabels
values = layer2_acc

# Create a horizontal bar plot
plt.figure(figsize=(16, 8))  # Adjust the figure size

# Choose colors for the bars
colors = sns.color_palette("inferno")

# Create the horizontal bar plot
bars = plt.barh(categories, values, color=colors)

# Add labels and title
plt.ylabel('Machine Learning Algorithms and Activation functions', fontsize= 12.0)
plt.xlabel('Accuracy in percentage', fontsize= 12.0)
plt.title('Accuracy comparison on 2 layers', fontsize= 16.0)
plt.xlim(0,110)

# Show the values on the right side of the bars
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2,
             f'{bar.get_width()}', ha='left', va='center')

# Show the plot
plt.show()
```
